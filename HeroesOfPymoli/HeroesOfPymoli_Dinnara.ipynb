```python
import pandas as pd

csv_file_path = "Resources/purchase_data.csv"

purchase_df = pd.read_csv(csv_file_path)
```

```python
players = purchase_df["SN"].unique()
total_players = len(players)
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_df["Item ID"].unique()
num_unique_items = len(unique_items)
average_price = purchase_df["Price"].mean()
unique_number_purchases = purchase_df['Purchase ID'].unique()
count_unique_number_purchases = len(unique_number_purchases)
total_revenue = purchase_df["Price"].sum()
total_purchase_df = pd.DataFrame({"Number of Unique Items":[num_unique_items],
                                  "Average Price":[average_price],
                                  "Number of Purchases":[count_unique_number_purchases],
                                  "Total Revenue":[total_revenue]})
total_purchase_df["Average Price"] = total_purchase_df["Average Price"].map("${:.2f}".format)
total_purchase_df["Total Revenue"] = total_purchase_df["Total Revenue"].map("${:,.2f}".format)
total_purchase_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_df = purchase_df.filter(['SN','Age','Gender'])
gender_df = gender_df.drop_duplicates()
total_gender_new = gender_df["Gender"].value_counts()
gender_demog_percent = (total_gender_new * 100)/ total_players
gender_demographics = pd.DataFrame({"Total Count": total_gender_new,
                                    "Percentage of Players": gender_demog_percent})
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}%".format)
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#missing Avg Total Purchase per Person


person_df = purchase_df.filter(['SN','Gender','Price'])


person_group = person_df.groupby(["SN","Gender"])
price_mean_person_group = person_group["Price"].mean()
purchase_analysis_temp = pd.DataFrame({"Price Avg per Person": price_mean_person_group})
print(purchase_analysis_temp)

gender_group_temp = purchase_analysis_temp.groupby(["Gender"])
average_temp = gender_group_temp["Price Avg per Person"].mean()
print(average_temp)

gender_group = person_df.groupby(["Gender"])
purchase_count = gender_group["SN"].count()
average_purchase_price = gender_group["Price"].mean()
total_purchase_value = gender_group["Price"].sum()


purchase_analysis = pd.DataFrame({"Purchase Count":purchase_count,
                                  "Average Purchase Price":average_purchase_price.map("${:.2f}".format),
                                 "Total Purchase Value":total_purchase_value.map("${:,.2f}".format),
                                 "Avg Total Purchase per Person":average_temp.map("${:,.2f}".format)})
purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
max_age = purchase_df.Age.max()
bins = [0,9.99,14,19,24,29,34,39,max_age+1]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
gender_df['Age Ranges'] = pd.cut(gender_df['Age'],bins=bins,labels=group_names)
age_ranges_group = gender_df.groupby(["Age Ranges"])
total_count = age_ranges_group["SN"].count()
percent_of_players = (total_count * 100) / total_players
age_demographics = pd.DataFrame({"Total Count": total_count,
                                "Percentage of Players": percent_of_players})
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}%".format)
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Missing Avg Total Purchase per Person
purchase_df['Age Ranges'] = pd.cut(purchase_df['Age'],bins=bins,labels=group_names)
pd_group_ages = purchase_df.groupby(["Age Ranges"])
average_price_age_range = pd_group_ages["Price"].mean()
purchase_count_age_range = pd_group_ages["Purchase ID"].count()
total_price_age_range = pd_group_ages["Price"].sum()


pd_group_ages_names = purchase_df.groupby(["SN","Age Ranges"])
total_price_ages_names_range = pd_group_ages_names["Price"].mean()

pd_group_ages_names_2 = purchase_df.groupby(["Age Ranges"])
total_price_ages_names_range_2 = pd_group_ages_names_2["Price"].mean()


purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count_age_range,
                                    "Average Purchase Price": average_price_age_range.map("${:.2f}".format),
                                   "Total Purchase Value":total_price_age_range.map("${:,.2f}".format),
                                   "Avg Total Purchase per Person":total_price_ages_names_range_2.map("${:,.2f}".format)})

purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_group = purchase_df.groupby(["SN"])
top_spenders_total_purchase = top_spenders_group["Price"].sum()
top_spenders_average_purchase_price = top_spenders_group["Price"].mean()
top_spenders_purchase_count = top_spenders_group["Purchase ID"].count()
top_spenders_demografics = pd.DataFrame({"Purchase Count": top_spenders_purchase_count,
                                    "Average Purchase Price": top_spenders_average_purchase_price,
                                   "Total Purchase Value":top_spenders_total_purchase})
top_spenders_sort = top_spenders_demografics.sort_values("Total Purchase Value", ascending=False).head(5)
top_spenders_sort["Average Purchase Price"]= top_spenders_sort["Average Purchase Price"].map("${:.2f}".format)
top_spenders_sort["Total Purchase Value"]= top_spenders_sort["Total Purchase Value"].map("${:.2f}".format)
top_spenders_sort
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_group = purchase_df.groupby(["Item ID","Item Name"])
most_pop_total_purchase = item_group["Price"].sum()
most_pop_average_price = item_group["Price"].mean()
most_pop_purchase_count = item_group["Price"].count()
most_popular_demografics = pd.DataFrame({"Purchase Count": most_pop_purchase_count,
                                         "Item Price": most_pop_average_price,
                                       "Total Purchase Value":most_pop_total_purchase})
most_popular_sort = most_popular_demografics.sort_values("Purchase Count", ascending=False).head(5)
most_popular_sort["Item Price"]= most_popular_sort["Item Price"].map("${:.2f}".format)
most_popular_sort["Total Purchase Value"]= most_popular_sort["Total Purchase Value"].map("${:.2f}".format)
most_popular_sort
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable_sort = most_popular_demografics.sort_values("Total Purchase Value", ascending=False).head(5)
most_profitable_sort["Item Price"]= most_profitable_sort["Item Price"].map("${:.2f}".format)
most_profitable_sort["Total Purchase Value"]= most_profitable_sort["Total Purchase Value"].map("${:.2f}".format)
most_profitable_sort
```
